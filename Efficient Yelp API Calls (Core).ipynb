```python
# Standard Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# additional Imports
import os, json, math, time
from yelpapi import YelpAPI
from tqdm.notebook import tqdm_notebook
```

```python
# Load API credentials
with open('/Users/pinl/.secret/yelp_api.json') as f:
    login = json.load(f)
login.keys()
```

```python
# Instantiate YelpAPI Variable
yelp_api = YelpAPI(login['API Key'],timeout_s=5.0)
```

```python
location = 'NY,NY 10118'
term = 'coffee'
```

```python
# specify folder for saving data
FOLDER = 'Data/'
os.makedirs(FOLDER,exist_ok=True)

# specifying json_file filename (can include a folder)
#JSON_FILE = 'New York-coffee'
JSON_FILE = FOLDER+f"{location.split(',')[0]}.json"
```

```python
JSON_FILE
```

```python
results = yelp_api.search_query(location=location,term=term)
```

```python
results.keys()
```

```python
pd.DataFrame(results['businesses'])
```

```python
# how many results total
total_results = results['total']
```

```python
total_results
```

```python
# page of results
results_per_page = len(results['businesses'])
results_per_page
```

```python
# Import additional packages for controlling our loop
import time, math
# Use math.ceil to round up for the total number of pages of results
n_pages = math.ceil((results['total']/results_per_page))
n_pages
```

```python
#results_per_page = None
#results_per_page
```

```python
## check if JSON FILE exists
if os.path.isfile(JSON_FILE)==False:
    ## print message if file does not exist and creat empty file
    print("The file does not exist. Creating empty file")
    with open(JSON_FILE,'w') as f:
        json.dump(results['businesses'],f)
## if file exists inform user
else:
    print('File exists')
```

```python
for i in tqdm_notebook( range(1,n_pages+1)):
    ## The block of code we want to TRY to run
    
    ## Read in results in progress file and check the length
    with open(JSON_FILE,'r') as f:
        previous_results = json.load(f)
    ## save number of results for to use as offset
    n_results = len(previous_results)
    
    if(n_results + results_per_page) > 1000:
        print('Exceeded 1000 api calls. Stopping loop.')
        break
    ## use n_results as the OFFSET
    results = yelp_api.search_query(location=location,
                                    term=term,
                                    offset=n_results)
    
    ## append new results and save to file
    previous_results.extend(results['businesses'])
    
    with open(JSON_FILE,'w') as f:
        json.dump(previous_results,f)
        
```

```python
# load final results
final_df = pd.read_json(JSON_FILE)
display(final_df.head(3),final_df.tail(3))
```

```python
# check for duplicate ID's 
final_df.duplicated(subset='id').sum()

```

```python
## Drop duplicate ids and confirm there are no more duplicates
final_df = final_df.drop_duplicates(subset='id')
final_df.duplicated(subset='id').sum()
```
